```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = [25, 30, 35, 40, 45]
series = pd.Series(data1, index=['A', 'B', 'C', 'D', 'E'])
print(series.head(3))
print(series.mean())
print(series.median())
print(series.std())
```

```python
data2 = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}
df1 = pd.DataFrame(data2)
print(df1.head(2))
print(df1.info())
df1['Passed'] = df1['Marks'] >= 80
print(df1)
```

```python
print(df1[['Name', 'Marks']])
print(df1[df1['Marks'] > 80])
print(df1.loc[df1['Marks'].idxmax()])
```

```python
df1.loc[1, 'Marks'] = None
df1.loc[4, 'Age'] = None
print(df1.isnull().sum())
df1['Marks'] = df1['Marks'].fillna(df1['Marks'].mean())
df1 = df1.dropna(subset=['Age'])
print(df1)
```

```python
grouped = df1.groupby('Gender')
print(grouped['Age'].mean())
print(grouped['Marks'].mean())
print(grouped.size())
```

```python
df1.to_csv('students_data.csv', index=False)
new_df = pd.read_csv('students_data.csv')
print(new_df.head(5))
```
